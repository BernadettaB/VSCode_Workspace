```python
%pip install ipykernel
# mini_chatbot.py
import difflib

#test 
#test2

# This is your "LLM brain": a simple dictionary
KNOWLEDGE = {
    "hello": "Hi there! ðŸ˜Š",
    "how are you": "I'm just Python code, but I'm running fine!",
    "what is your name": "I'm a tiny dictionary-based chatbot.",
    "bye": "Goodbye! ðŸ‘‹"
}

def get_best_match(user_input: str, threshold: float = 0.4):
    """
    Find the closest matching key in KNOWLEDGE using fuzzy matching.
    Returns (best_key, score) or (None, 0) if nothing is close enough.
    """
    user_input = user_input.lower().strip()
    best_key = None
    best_score = 0.0

    for key in KNOWLEDGE.keys():
        score = difflib.SequenceMatcher(None, user_input, key).ratio()
        if score > best_score:
            best_score = score
            best_key = key

    if best_score >= threshold:
        return best_key, best_score
    return None, 0.0

def chat():
    print("Dictionary Chatbot ðŸ¤– (type 'quit' to exit)")
    while True:
        user_input = input("You: ").strip()
        if user_input.lower() in {"quit", "exit", "bye"}:
            print("Bot: Bye! ðŸ‘‹")
            break

        key, score = get_best_match(user_input)
        if key is not None:
            print(f"Bot: {KNOWLEDGE[key]}  (matched: '{key}', score={score:.2f})")
        else:
            print("Bot: Sorry, I don't know how to answer that yet.")

if __name__ == "__main__":
    chat()
```
